```python
#Se importan las librerias y paquetes
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([1,2,3,4,5]).reshape(-1,1) #variables independientes
y = np.array([1.5,3,2.8,4.3,6.1]) #variables dependientes
```

```python
#Se crea el modelo de RegresiÃ³n Lineal
modelo = LinearRegression()
modelo.fit(x, y)
y_pred = modelo.predict(x) #se hacen prediccones utilizando el modelo anterior
```

```python
#Se grafican los datos y la linea de la regresiÃ³n
plt.scatter(x, y, color='blue', label='Datos')
plt.plot(x, y_pred, color='red', linewidth=2, label='RegresiÃ³n Lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
#Se muestra la pendiente y el intercepto del modelo
print(f'Coeficiente (pendiente): {modelo.coef_[0]}')
print(f'Intercepto: {modelo.intercept_}')
```

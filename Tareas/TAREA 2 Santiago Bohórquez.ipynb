La **regresión lineal** es una técnica de análisis de datos que predice el valor de datos desconocidos mediante el uso de otro valor de datos relacionado y conocido. Esta técnica se puede aplicar fácilmente en la computación en ámbitos como el *machine learning* y la *inteligencia artificial*. En la ciencia de datos la regresión lineal es uno de los algoritmos más utilizados para diferentes aplicaciones como:
1. *Predicción:* Donde se utiliza la regresión lineal para predecir valores númericos en función de variables independientes
2. *Análisis de tendencias:* Se utiliza para analizar tendencias en datos a lo largo de una serie de timepo, para identidficar patrones o cambios.
3. *Evaluación de Relaciones Causales:* Permite analizar relaciones causa-efecto entre variables.

La regresión lineal en esencia, es una técnica que intenta trazar un gráfico lineal entre 2 variables de datos "x & y". X como la variable **independiente** (también denominada variable explicativa o variable rpredictiva) se traza a lo largo del eje horizontal. La variable **dependiente** y, se traza en el eje vertical.

Existen dos tipos de regresión lineal:
1. La Regresión Lineal Simple
2. La Regresión Lineal Múltiple

##Regresión Lineal Simple

Se define mediante la siguiente ecuación:
**Y = β0X + β1 + ε** donde,
β0 y β1 son dos constantes desconocidas que representan la pendiente de la regresión, y **ε (épsilon)** que es el término de error.

##Regresión Lineal Múltiple

En la regresión múlyiple el conjunto de datos contiene una variable dependiente y múltiples variables independientes. Se define de la siguiente manera:
**Y = β0x0 + β1x1 + β2x2+…… βNxN+ ε**
En este caso las constantes β aumentan a medida que el número de variables predictivas también aumentan.

```python
#Se importan las librerias y paquetes
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([1,2,3,4,5]).reshape(-1,1) #variables independientes
y = np.array([1.5,3,2.8,4.3,6.1]) #variables dependientes
```

```python
#Se crea el modelo de Regresión Lineal
modelo = LinearRegression()
modelo.fit(x, y)
y_pred = modelo.predict(x) #se hacen prediccones utilizando el modelo anterior
```

```python
#Se grafican los datos y la linea de la regresión
plt.scatter(x, y, color='blue', label='Datos')
plt.plot(x, y_pred, color='red', linewidth=2, label='Regresión Lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
#Se muestra la pendiente y el intercepto del modelo
print(f'Coeficiente (pendiente): {modelo.coef_[0]}')
print(f'Intercepto: {modelo.intercept_}')
```
